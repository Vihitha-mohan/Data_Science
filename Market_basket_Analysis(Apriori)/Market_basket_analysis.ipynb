```python
import pandas as pd
df = pd.read_csv('bread_basket.csv')
df.head()
```

```python
df.info()
```

```python
transactions = df.groupby('Transaction')['Item'].apply(list).to_list()
transactions
```

```python
from mlxtend.preprocessing import TransactionEncoder

encoder = TransactionEncoder()
onehot = encoder.fit(transactions).transform(transactions)
onehot_df = pd.DataFrame(onehot, columns=encoder.columns_)
onehot_df
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(onehot_df, min_support=0.05, use_colnames=True)
print(frequent_itemsets)
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
print(rules)
```

```python
#Rules
for index, rule in rules.iterrows():
    print(f"Rule: {rule['antecedents']} -> {rule['consequents']} (Support: {rule['support']:.4f}, Confidence: {rule['confidence']:.4f}, Lift: {rule['lift']:.4f})")
```
